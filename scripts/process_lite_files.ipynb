# Process OCO-2 Lite files for SIF and XCO2 datasets

```python
import sys
sys.path.insert(0, "../src")

import xarray
import data_utils
```

```python
# HPC: Open datasets for all years (2014-09-06:2020-07-31)
paths_sif = "/data/OCO-2/OCO2_L2_Lite_SIF.10r/*/*.nc4"
paths_xco2 = "/data/OCO-2/OCO2_L2_Lite_FP.10r/*/*.nc4"

DS_sif = xarray.open_mfdataset(paths_sif, preprocess=data_utils.prep_sif, parallel=True)
DS_xco2 = xarray.open_mfdataset(paths_xco2, preprocess=data_utils.prep_xco2, parallel=True)
```

```python
# Write datasets to smaller netcdfs
DS_sif.to_netcdf("../data/exp_pro/OCO2_Lite_SIF740.nc", format="NETCDF4")
DS_xco2.to_netcdf("../data/exp_pro/OCO2_Lite_XCO2.nc", format="NETCDF4")
```
